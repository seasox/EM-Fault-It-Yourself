```python

import json
import matplotlib.pyplot as plt 
```

```python

fname = "../dp_json/bikel1_2024-04-15T00:05:02.966748.json"
```

```python
with open(fname, "r") as fp:
    data = json.load(fp)
experiments = data['experiments']
settings = data['settings']
print(settings)

weights = [experiment['hw_h0'] if 'hw_h0' in experiment else None for experiment in experiments]
print(weights)

fig = plt.figure(figsize = (10, 5))

# creating the bar plot
plt.bar([i for i in range(len(weights))], weights, width = 0.6)
    
plt.xlabel("Experiment #")
plt.ylabel("H0 Weight")
plt.title("BIKE L1 H0 Weights")
plt.show()
```
